### Exercise 6

```python
locations = ['USA', 'Dubai', 'Canada', 'Egypt', 'Siwa']
locations
```

```python
# to sort the locations alphabetically without modifying the original list
print(sorted(locations))
# to print the original list
print(locations)
```

```python
# to sort list in reverse alphabetical order modifying of the original list.
print(sorted(locations, reverse = True))
```

```python
# check that the original list hasn't changed
print(locations)
```

```python
# to reverse the order of the original list
locations.reverse()
print(locations)
```

```python
# to reverse back to the original list
locations.reverse()
print(locations)
```

```python
# to sort the locations alphabetically with modifying the original list
locations.sort()
print(locations)
```

```python
# to sort list in reverse alphabetical order modifying of the original list
locations.sort(reverse=True)
print(locations)
```
