### Exercise 9

```python
class Restaurant:
    def __init__(self, restaurant_name, cuisine_type):
        self.restaurant_name = restaurant_name
        self.cuisine_type = cuisine_type
        self.open = False
        self.num_served = 0
        
    def open_restaurant(self):
        self.open = True
    def close_restaurant(self):
        self.open = False
        
    def describe_restaurant(self):
        print(f" {self.restaurant_name.title()} serves {self.cuisine_type.title()} food")
        
    def is_open(self):
        if self.open == True: 
            print(f" {self.restaurant_name.title()} is Open") 
        else:
            print(f" {self.restaurant_name.title()} is Closed")  
            
    def set_num_served(self, number):
        self.num_served = number
    
    def increment_num_served(self):
        self.num_served += 1
    
```

```python
italian = Restaurant("Miko", "Italian")
greek = Restaurant("Yani", "Japanise")
```

```python
greek.set_num_served(92)
print(greek.num_served)
```

```python
italian.increment_num_served()
print(italian.num_served)
```

### Exercise 10

```python
class IceCreamStand(Restaurant):
    def __init__(self,  restaurant_name, cuisine_type, flavors):
        super().__init__(restaurant_name, cuisine_type)
        self.flavors = flavors
        
    def display_flavors(self):
        for flavor in self.flavors:
            print(f" we have {flavor}!")
```

```python
my_icecream = IceCreamStand("Vona", "Ice Cream", ['Mango', 'Vanilia', 'Strawberry'])
```

```python
my_icecream.display_flavors()
```

```python
my_icecream.set_num_served(62)
print(my_icecream.num_served)
```

```python
my_icecream.open_restaurant()
my_icecream.is_open()
```
